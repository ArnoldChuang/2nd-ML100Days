```python
# 做完特徵工程前的所有準備 (與前範例相同)
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression

data_path = '../data/'
df_train = pd.read_csv(data_path + 'house_train.csv.gz')

train_Y = np.log1p(df_train['SalePrice'])
df = df_train.drop(['Id', 'SalePrice'] , axis=1)
df.head()
```

```python
#只取 int64, float64 兩種數值型欄位, 存於 num_features 中
num_features = []
for dtype, feature in zip(df.dtypes, df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
print(f'{len(num_features)} Numeric Features : {num_features}\n')
```

```python
# 削減文字型欄位, 只剩數值型欄位
df = df[num_features]
df = df.fillna(-1)
# MinMaxScaler(), 將資料做最大最小標準化
MMEncoder = MinMaxScaler()
train_num = train_Y.shape[0]
df.head()
```

# 作業1
* 試著限制 '1樓地板面積(平方英尺)' (1stFlrSF) 欄位的上下限, 看看能否再進一步提高分數?

```python
# 顯示 1stFlrSF 與目標值的散佈圖
import seaborn as sns
import matplotlib.pyplot as plt
sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
# 先將資料 df 做標準化
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
# 搭配線性回歸, 使用 cross_val_score 做交叉驗證 和 結果評估
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 調整離群值
#clip函數將 群組值限制在 a_min, a_max之間，大於a_max的元素就等於 a_max，小於 a_min 的元素就等於 a_min

#df['1stFlrSF'][:train_num] = np.clip(df['1stFlrSF'][:train_num] ,350,3500)
#sns.regplot(x = df['1stFlrSF'][:train_num], y=train_Y)
df['1stFlrSF'] = df['1stFlrSF'].clip(500,2500)
sns.regplot(x = df['1stFlrSF'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

# 作業2
* 續前題, 去除離群值有兩類方式 :  捨棄離群值(刪除離群的資料) 以及調整離群值,  
請試著用同樣的上下限, 改為 '捨棄離群值' 的方法, 看看結果會變好還是變差? 並試著解釋原因。

```python
# 將 1stFlrSF 限制在你覺得適合的範圍內, 捨棄離群值
# 將在範圍內的 index 記錄下來
keep_indexs = (df['1stFlrSF']> 500) & (df['1stFlrSF']< 2500)
keep_indexs.head()
```

```python
# df = df_train.drop(['Id', 'SalePrice'] , axis=1)
# df 只保留 keep_indexs 的對應資料
df = df[keep_indexs]
df.head()
```

```python
# train_Y = np.log1p(df_train['SalePrice'])
# train_Y 只保留 keep_indexs 的對應資料
train_Y = train_Y[keep_indexs]
train_Y.head()
```

```python
sns.regplot(x = df['1stFlrSF'], y=train_Y)
plt.show()

# 做線性迴歸, 觀察分數
train_X = MMEncoder.fit_transform(df)
estimator = LinearRegression()
cross_val_score(estimator, train_X, train_Y, cv=5).mean()
```

ans: 捨棄離群值的結果看似比較好, 因不合理的值變少了, 所以線性回歸的分數提高了
