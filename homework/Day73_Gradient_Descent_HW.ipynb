    主要驗證 Lr 對於grandient 收斂的速度
    

```python
precision = 0.000001 #This tells us when to stop the algorithm
df = lambda x: 2*(x+5) #Gradient of our function
```

```python
def iteration_while(cur_x, iters_history, x_history, lr, iters, previous_step_size):
    while previous_step_size > precision and iters < max_iters:
        prev_x = cur_x #Store current x value in prev_x
        cur_x = cur_x - lr * df(prev_x) #Gradient descent
        previous_step_size = abs(cur_x - prev_x) # 取較大的值, Change in x
        iters = iters+1 #iteration count
        #print("Iteration",iters,"\nX value is",cur_x) #Print iterations
        # Store parameters for plotting
        iters_history.append(iters)
        x_history.append(cur_x)
    print("the learning rate is",lr, "\nIteration",iters,"\nX value is",cur_x) #Print iterations

# initial param, Learning rate = 1.0
cur_x = 3
# Learning rate
lr = 1.0 
#iteration counter
iters = 0
# maximum number of iterations
max_iters = 10000 
# record history
iters_history_1 = [iters]
x_history_1 = [cur_x]
#Step size 
previous_step_size = 1  
iteration_while(cur_x, iters_history_1, x_history_1, lr, iters, previous_step_size)

# initial param, Learning rate = 0.1
cur_x = 3
# Learning rate
lr = 0.1 
#iteration counter
iters = 0
# maximum number of iterations
max_iters = 10000
# record history
iters_history_01 = [iters]
x_history_01 = [cur_x]
#Step size 
previous_step_size = 1  
iteration_while(cur_x, iters_history_01, x_history_01, lr, iters, previous_step_size)

# initial param, Learning rate = 0.01   
cur_x = 3 
# Learning rate
lr = 0.01 # Learning rate
#iteration counter
iters = 0
# maximum number of iterations
max_iters = 10000
# record history
iters_history_001 = [iters]
x_history_001 = [cur_x]
#Step size 
previous_step_size = 1     
# Call iteration_while run result
iteration_while(cur_x, iters_history_001, x_history_001, lr, iters, previous_step_size)

# initial param, Learning rate = 0.001
cur_x = 3
# Learning rate
lr = 0.001 
#iteration counter
iters = 0
# maximum number of iterations
max_iters = 10000
# record history
iters_history_0001 = [iters]
x_history_0001 = [cur_x]
#Step size 
previous_step_size = 1  
iteration_while(cur_x, iters_history_0001, x_history_0001, lr, iters, previous_step_size)

# initial param, Learning rate = 0.0001
cur_x = 3
# Learning rate
lr = 0.0001 
#iteration counter
iters = 0
# maximum number of iterations
max_iters = 100000
# record history
iters_history_00001 = [iters]
x_history_00001 = [cur_x]
#Step size 
previous_step_size = 1  
iteration_while(cur_x, iters_history_00001, x_history_00001, lr, iters, previous_step_size)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
#適用於 Jupyter Notebook, 宣告直接在cell 內印出執行結果

def Draw_Plt(lr, iters_history, x_history):
    plt.title('the learning rate is ' + str(lr) )
    plt.plot(iters_history, x_history, 'o-', ms=3, lw=1.5, color='black')
    plt.xlabel(r'$iters$', fontsize=16)
    plt.ylabel(r'$x$', fontsize=16)
    plt.show()


Draw_Plt(1.0, iters_history_1, x_history_1)
Draw_Plt(0.1, iters_history_01, x_history_01)
Draw_Plt(0.01, iters_history_001, x_history_001)
Draw_Plt(0.001, iters_history_0001, x_history_0001)
Draw_Plt(0.0001, iters_history_00001, x_history_00001)
```
