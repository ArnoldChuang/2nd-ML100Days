```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
#隨機產生 0.0 ~ 1.0的亂數,大小為 10x10
#matrix = np.random.rand(10,10)

#uniform,生成一個指定範圍內的隨機浮點數，low為下限，high為上限
matrix = np.random.uniform(low=-1, high=1, size=(10,10))
plt.figure(figsize=(10,10))
"""
Your Code Here
"""
# 繪製 0.1 ~ 1.0 的 Heatmap
plt.title('-1.0 ~ 1.0 Heatmap');
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.show()
```

```python
#隨機產生 50以下的亂數整數,大小為 8x6
matrix = np.random.randint(50, size = (8,6))
plt.figure(figsize=(10,10))

# Random int50 的 Heatmap
plt.title('Random int50 Heatmap');
heatmap = sns.heatmap(matrix, cmap = plt.cm.RdYlBu_r, vmin = -0.25, annot = True, vmax = 0.6)
plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
#uniform,生成一個指定範圍內的隨機浮點數，low為下限，high為上限
matrix = np.random.uniform(low=-1, high=1, size=(nrow,ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, 
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
#依據上面的 grid 資料使用map繪製以下圖表
# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)
# 對角線畫 histogram
grid.map_diag(sns.kdeplot)
# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
#loc：float,此機率分布的均值（loc=0 意即以 x=0 為中心centre）
#scale：float,標準差值（scale越大分布圖越矮胖，scale越小越瘦高）
#size：int or tuple of ints, 输出的shape，預設為None只輸出一個值
matrix = np.random.normal(loc=2.0, scale=1.5, size=(nrow,ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice).reset_index()

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False,
                    hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
"""
Your Code Here
Please replace "..." to correct plot function
"""
# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)
# 對角線畫 histogram
grid.map_diag(sns.kdeplot)
# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
